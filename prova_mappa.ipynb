```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
import scipy.special as sc
import pandas as pd
import statsmodels.api as sm
import scipy.stats as st
import math
import csv
import plotly.express as px
```

```python
air_quality= pd.read_csv('air_quality/AQI_lat_long.csv', sep=',', decimal='.')
ita = air_quality[air_quality['Country'] == 'Italy']
switz = air_quality[air_quality['Country'] == 'Switzerland']
merged = ita.merge(switz, how='outer')
```

```python
fig = px.density_mapbox(merged, lat = 'lat', lon = 'lng', z = 'AQI Value', radius = 20,
                        center = dict(lat = merged.lat.mean(), lon = merged.lng.mean()),
                        zoom = 5, mapbox_style = 'carto-positron', height = 900, opacity=0.5, title = 'Air Quality Value')
fig.show()
```
