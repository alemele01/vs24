```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
import scipy.special as sc
import pandas as pd
import statsmodels.api as sm
import scipy.stats as st
import math
import csv
```

```python
statista = pd.read_csv('statista.csv')
statista.sort_values(by="Percent", ascending = True, inplace= True)
statista.head()
```

```python
bars = plt.barh(statista['Country'], statista['Percent'])
plt.title('People Satisfied with public transport in referred Country')
plt.rcParams["figure.figsize"] = (10,15)
plt.grid(axis='x', linestyle = '--')
plt.show()
```

```python
statista_it = statista[statista['Country'] == "Italy"]
statista_sw = statista[statista['Country'] == "Switzerland"]
statista_uk = statista[statista['Country'] == "United Kingdom"]
statista_ger = statista[statista['Country'] == "Germany"]
statista_fr = statista[statista['Country'] == "France"]
statista_eu = pd.DataFrame()
statista_eu = pd.concat([statista_eu, statista_it])
statista_eu = pd.concat([statista_eu, statista_sw])
statista_eu = pd.concat([statista_eu, statista_uk])
statista_eu = pd.concat([statista_eu, statista_ger])
statista_eu = pd.concat([statista_eu, statista_fr])
statista_eu.head()
```

```python
plt.barh(statista_eu['Country'], statista_eu['Percent'])
plt.title('People Satisfied with public transport in referred Country')
plt.rcParams["figure.figsize"] = ()
plt.grid(axis='x', linestyle = '--')
plt.show()
```
